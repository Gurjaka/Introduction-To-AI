## კლასების შესავალი
ისწავლეთ ობიექტზე ორიენტირებული პროგრამირების საფუძვლები კლასების გამოყენებით.

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        return f"Hello, my name is {self.name} and I am {self.age} years old."

person = Person("Alice", 25)
print(person.greet())
````

### სავარჯიშო
შექმენით კლასი `Car`, რომელსაც ექნება თვისებები `brand` და `model`, და მეთოდი `description`, რომელიც დააბრუნებს ტექსტს: "This car is a [brand] [model]." შექმენით ობიექტი და დაბეჭდეთ მისი აღწერა.

````python
class Car:
  def __init__(self, brand: str, model: str) -> None:
    self.brand = brand
    self.model = model

  def description(self) -> str:
    return f"This car is a {self.brand} {self.model}"


x = Car("Tesla", "Juniper")
print(x.description())
````
