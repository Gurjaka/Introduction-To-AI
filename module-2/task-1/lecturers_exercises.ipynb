### áƒ¡áƒáƒ•áƒáƒ áƒ¯áƒ˜áƒ¨áƒ
áƒ¨áƒ”áƒ¥áƒ›áƒ”áƒœáƒ˜áƒ— áƒ™áƒšáƒáƒ¡áƒ˜ `Car`, áƒ áƒáƒ›áƒ”áƒšáƒ¡áƒáƒª áƒ”áƒ¥áƒœáƒ”áƒ‘áƒ áƒ—áƒ•áƒ˜áƒ¡áƒ”áƒ‘áƒ”áƒ‘áƒ˜ `brand` áƒ“áƒ `model`, áƒ“áƒ áƒ›áƒ”áƒ—áƒáƒ“áƒ˜ `description`, áƒ áƒáƒ›áƒ”áƒšáƒ˜áƒª áƒ“áƒáƒáƒ‘áƒ áƒ£áƒœáƒ”áƒ‘áƒ¡ áƒ¢áƒ”áƒ¥áƒ¡áƒ¢áƒ¡: "This car is a [brand] [model]." áƒ¨áƒ”áƒ¥áƒ›áƒ”áƒœáƒ˜áƒ— áƒáƒ‘áƒ˜áƒ”áƒ¥áƒ¢áƒ˜ áƒ“áƒ áƒ“áƒáƒ‘áƒ”áƒ­áƒ“áƒ”áƒ— áƒ›áƒ˜áƒ¡áƒ˜ áƒáƒ¦áƒ¬áƒ”áƒ áƒ.

````python
class Car:
  def __init__(self, brand: str, model: str) -> None:
    self.brand = brand
    self.model = model

  def description(self) -> str:
    return f"This car is a {self.brand} {self.model}"


x = Car("Tesla", "Juniper")
print(x.description())
````
