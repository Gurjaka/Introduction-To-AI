9) If Statements

```python
temperature = 25

if temperature > 30:
  print("It's a hot day")
  print("Drink plenty of water")
elif temperature > 20:
  print("It's a nice day")
elif temperature > 10:
  print("It's a bit cold")
print("Done")
```

10) Exercise

```python
weight = int(input("Weight: "))
unit = input("(K)g or (L)bs: ")

if unit.upper() == "K":
    converted = weight / 0.45
    print("Weight in Lbs: " + str(converted))
else:
    converted = weight / 0.45
    print("Weight in Kgs: " + str(converted))
```

```python
numbers = [1, 2, 3, 4, 5]

for item in numbers:
    print(item)

i = 0
while i < len(numbers):
    print(numbers[i])
    i += 1

```
